## Data Cleaning Stage
Since the dataset contains missing values I need to handle them for each columns using the following aproaches:
- **`Age`**: Imputing missing values using the mean.
- **`Fare`**: Single missing value can be imputed using the median fare.
- **`Cabin`**: The high percentage of missing values (78.2%) suggests that this column may not be useful for analysis unless additional information is available.

```python
age_mean: float = df['Age'].mean().__round__(2)
age_mean
```

```python
df['Age'].fillna(age_mean, inplace=True)
df
```

```python
median_fare: float = df['Fare'].median()
median_fare
```

```python
df['Fare'].fillna(median_fare, inplace=True)
df
```

```python
df.drop(columns='Cabin', inplace=True)
df
```
